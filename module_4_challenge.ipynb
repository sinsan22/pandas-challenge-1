## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df["subtotal"] = df["unit_price"] * df["qty"]
df.head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

df['shipping_price'] = df.apply(lambda row: 7 * row['unit_weight'] * row['qty'] if row['unit_weight'] * row['qty']> 50 else 10 * row['unit_weight'] * row['qty'], axis=1)
df.head(3)

```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
# created a variable for with sales tax converted for multiplication 
sales_tax_rate = .0925 
# variable to store the actual amount of the sales tax
#sales_tax_amt = df['subtotal'] * sales_tax_rate
sales_tax_amt = (df['subtotal'] + df['shipping_price']) * sales_tax_rate
df['total_price'] = df['subtotal'] + df['shipping_price'] + sales_tax_amt

# Rounded to 2 decimal places
df['total_price'] = df['total_price'].round(2)

df.head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']

df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['profit'] = df['total_price'] - df['cost']
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

for client_id, qty in top_five.items():
    total_spent = df[df['client_id'] == client_id]['total_price'].sum()
    
    print(f"Client {client_id} ordered {qty} and spent {total_spent: .2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# My original code reference the top_five as a data frame when it was a series. After 2 weeks of struggle, I ran the error through ChatGPT and finally 
#figured out the issue. ChaptGPT helped me with the code to reference the top_five_list.

top_clients = df[df['client_id'].isin(top_five_list)]

#This code groups by client_id and calculated totals
totals_df = top_clients.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'profit': 'sum'

}).reset_index()

# Code sorts by total profit in descending order
totals_df = totals_df.sort_values(by = 'profit', ascending = False)

print(totals_df)

```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
# Get a list of column names
# df.columns
new_column_names = ['First Name', 'Last Name', 'Job Title', 'Phone', 'Email', 
                    'Client ID', 'Order ID', 'Order Date', 'Order Week', 
                    'Order Year', 'Item ID', 'Category', 'Subcategory', 'Unit Price',
                    'Unit Cost', 'Unit Weight', 'Quantity', 'Line Number', 'Subtotal', 
                    'Shipping Price', 'Total Price', 'Cost', 'Profit']
df.columns = new_column_names
# Checking new column names
# df.columns
updated_currency = ['Subtotal', 'Shipping Price', 'Total Price', 'Cost', 'Profit']
df[updated_currency] = df[updated_currency] / 1000000

df = df.round(2)

for col in updated_currency:
    df[col] = df[col].apply(lambda x: f"${x:.2f}M")



df.head(10)
```
